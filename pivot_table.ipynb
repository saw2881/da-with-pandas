- index - A column or a list of columns to group the data on
- values - The column to aggregate on
- aggfunc (optional) - The aggregation function such as min, max, count, mean, sum etc to apply (defaults to "mean")
- columns (optional) - The columns to display

```python
import pandas as pd
foods = pd.read_csv("data/foods.csv")
foods
```

```python
# "aggfunc" defaults to mean
foods.pivot_table(index="Gender", values="Spend")
```

```python
foods.pivot_table(index="Gender", values="Spend", aggfunc="sum")
```

```python
# grouped by "Gender" and then by "Item"
foods.pivot_table(index=["Gender", "Item"], values="Spend")
```

```python
# three layers of grouping Gender / Item / City
foods.pivot_table(index=["Gender", "Item", "City"], values="Spend")
```

```python
# a columunar represenstation
foods.pivot_table(index=["Gender", "Item"], columns="City", values="Spend")
```
